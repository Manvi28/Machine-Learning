```python
# ==============================
# Linear Regression with scikit-learn
# Category 2: Using ready-made ML library
# ==============================

# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, SGDRegressor

# ------------------------------
# Step 1: Generate synthetic dataset
# ------------------------------
np.random.seed(42)                               # fix seed for reproducibility
X = 2 * np.random.rand(100, 1)                   # 100 random values for X in range [0,2]
y = 4 + 3 * X + np.random.randn(100, 1)          # true relation y = 4 + 3x + noise

# Plot the data
plt.scatter(X, y, color="blue", label="Data Points")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Generated Data for Linear Regression (scikit-learn)")
plt.legend()
plt.show()
```

```python

# ------------------------------
# Step 2: LinearRegression (closed-form solution internally)
# ------------------------------
lin_reg = LinearRegression()                     # create model object
lin_reg.fit(X, y)                                # train model on dataset

print("LinearRegression Results:")
print("Intercept (theta0):", lin_reg.intercept_)
print("Slope (theta1):", lin_reg.coef_)

# Predictions for new values
X_new = np.array([[0], [2]])
y_pred_lin = lin_reg.predict(X_new)

# Plot regression line
plt.scatter(X, y, color="blue", label="Data Points")
plt.plot(X_new, y_pred_lin, "g-", linewidth=2, label="LinearRegression Line")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression using scikit-learn")
plt.legend()
plt.show()




```

```python
# ------------------------------
# Step 3: SGDRegressor (gradient descent internally)
# ------------------------------
sgd_reg = SGDRegressor(max_iter=1000,            # max passes over data
                       tol=1e-3,                 # stop if improvement is very small
                       penalty=None,             # no regularization
                       eta0=0.1,                 # learning rate
                       random_state=42)

sgd_reg.fit(X, y.ravel())                        # .ravel() converts y into 1D array

print("\nSGDRegressor Results:")
print("Intercept (theta0):", sgd_reg.intercept_)
print("Slope (theta1):", sgd_reg.coef_)

# Predictions for new values
y_pred_sgd = sgd_reg.predict(X_new)

# Plot regression line
plt.scatter(X, y, color="blue", label="Data Points")
plt.plot(X_new, y_pred_sgd, "orange", linewidth=2, label="SGDRegressor Line")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression using SGDRegressor")
plt.legend()
plt.show()
```

```python
# ------------------------------
# Step 4: Compare both methods on same graph
# ------------------------------
plt.scatter(X, y, color="blue", label="Data Points")
plt.plot(X_new, y_pred_lin, "g-", linewidth=2, label="LinearRegression")
plt.plot(X_new, y_pred_sgd, "orange", linewidth=2, linestyle="--", label="SGDRegressor")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Comparison: LinearRegression vs SGDRegressor")
plt.legend()
plt.show()
```
